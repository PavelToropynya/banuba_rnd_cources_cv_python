```python
import numpy as np
import matplotlib.pyplot as plt
import os
import random
import cv2
from PIL import Image, ImageOps
```

```python
height = 505
width = 325
image_orig = cv2.imread('phone.jpeg')[:, :, ::-1]
plt.imshow(image_orig)
```

```python
image_changed = image_orig.copy()
black = 5
white = 255
gray = 100
image_changed[np.logical_and(image_changed < white, image_changed > black)] = gray
plt.imshow(image_changed)
```

```python

image_all_pixels = image_changed.ravel()
print(image_all_pixels.size)

background_size = len(list(filter(lambda x: x == white, image_all_pixels)))#[a for a in image_all_pixels if a in white_set]
print(background_size)
phone_screen_size = len(list(filter(lambda x: x == gray, image_all_pixels)))#[a for a in image_all_pixels if a in gray_set]
print(phone_screen_size)
phone_size = image_all_pixels.size - background_size
print(phone_size)
screen_percent = phone_screen_size *100 / phone_size
print(screen_percent)
```
